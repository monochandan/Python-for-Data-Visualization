```python
data['CO(GT)'] = data['CO(GT)'].mask(data['CO(GT)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['PT08.S1(CO)'] = data['PT08.S1(CO)'].mask(data['PT08.S1(CO)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['NMHC(GT)'] = data['NMHC(GT)'].mask(data['NMHC(GT)'].lt(0.0)).ffill().fillna(0.0).astype('int64')
data['C6H6(GT)'] = data['C6H6(GT)'].mask(data['C6H6(GT)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['PT08.S2(NMHC)'] = data['PT08.S2(NMHC)'].mask(data['PT08.S2(NMHC)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['NOx(GT)'] = data['NOx(GT)'].mask(data['NOx(GT)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['PT08.S3(NOx)'] = data['PT08.S3(NOx)'].mask(data['PT08.S3(NOx)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['NO2(GT)'] = data['NO2(GT)'].mask(data['NO2(GT)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['PT08.S4(NO2)'] = data['PT08.S4(NO2)'].mask(data['PT08.S4(NO2)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['PT08.S5(O3)'] = data['PT08.S5(O3)'].mask(data['PT08.S5(O3)'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['T'] = data['T'].mask(data['T'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['RH'] = data['RH'].mask(data['RH'].lt(0.0)).ffill().fillna(0.0).astype('float64')
data['AH'] = data['AH'].mask(data['AH'].lt(0.0)).ffill().fillna(0.0).astype('float64')
```
